```python
import networkx as nx
%matplotlib inline
```

```python
M=nx.read_gml("polblogs.gml")
```

```python
# create weighted graph from M
G = nx.Graph()
for u,v,data in M.edges_iter(data=True):
    if G.has_edge(u,v):
        G[u][v]['weight'] += 1
    else:
        G.add_edge(u, v, weight=1)

```

```python
centrality = nx.eigenvector_centrality_numpy(G)
```

```python
centrality
```

```python
for key, val in centrality.iteritems():
    centrality[key] = round(val, 5)
```

```python
centrality
```

### 1.  Create a nice looking graph of the political blog network

```python
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(40,40))
pos = nx.spring_layout(G, scale=5)
nx.draw_networkx_labels(G,pos, centrality, font_size=7)
nx.draw_networkx_nodes(G, pos, node_size=500, alpha=0.6, node_color='w')
nx.draw_networkx_edges(G, pos)
plt.axis('off')
plt.show()
```

```python
influential = []
for node in centrality:
    influential.append((node,centrality[node]))
```

```python
sorted(influential, key=lambda x : x[1], reverse=True)
```
